```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
sort = pd.read_csv('sort.txt')
```

```python
sort = list(sort)
sort_array = [float(i) for i in sort[0:11]]
add_hash = [float(i) for i in sort[11:22]]
add_tree = [float(i) for i in sort[22:33]]
```

```python
plt.figure(figsize=(15, 9))
grid = np.arange(100, 10001, 990)
plt.plot(grid, sort_array, label='Array')
plt.plot(grid, add_hash, label='HashTable')
plt.plot(grid, add_tree, label='AVLTree')
plt.title('График зависимости времени размещения данных в структурах от количества элеметов')
plt.legend()
plt.show()
```

Как и ожидалось, с увеличением количества элементов растет время выполнение операции. Это мы будем наблюдать для всех операций. Как можно видеть, сортировка массива происходит быстрее, чем последовательная вставка в дерево или хэш-таблицу.

```python
find = pd.read_csv('find.txt')
find = list(find)
find_array = [float(i) for i in find[0:11]]
find_hash = [float(i) for i in find[11:22]]
find_tree = [float(i) for i in find[22:33]]
```

```python
plt.figure(figsize=(15, 9))
grid = np.arange(100, 10001, 990)
plt.plot(grid, find_array, label='Array')
plt.plot(grid, find_hash, label='HashTable')
plt.plot(grid, find_tree, label='AVLTree')
plt.title('График зависимости времени поиска исходных данных в структурах от количества элеметов')
plt.legend()
plt.show()
```

Снова видим, что двоичный поиск в массиве оказался немного быстрее, чем в дереве, и много быстрее, чем в хэш-таблице.

```python
delete = pd.read_csv('delete.txt')
delete = list(delete)
delete_array = [float(i) for i in delete[0:11]]
delete_hash = [float(i) for i in delete[11:22]]
delete_tree = [float(i) for i in delete[22:33]]
```

```python
plt.figure(figsize=(15, 9))
grid = np.arange(100, 10001, 990)
plt.plot(grid, delete_array, label='Array')
plt.plot(grid, delete_hash, label='HashTable')
plt.plot(grid, delete_tree, label='AVLTree')
plt.title('График зависимости времени удаления данных из структур от количества элеметов')
plt.legend()
plt.show()
```

А вот удаление массива работает значительно дольше, чем очистка хэш-таблицы и дерева.

При сравнении времени поэлементного удаления элементов из хэш-таблицы и дерева видим, что дерево удаляется быстрее.
